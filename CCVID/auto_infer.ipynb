```python
import os

gallery_splits_dir = "/home/ika/yzlm/TwinProject/CCVID/data/bounding_box_test_split"
query_splits_dir = "/home/ika/yzlm/TwinProject/CCVID/data/query_splits"
yaml_dir = "/home/ika/yzlm/TwinProject/CCVID/train_with_val.yaml"
model_dir = "/home/ika/yzlm/TwinProject/CCVID/results/train/model_epoch_029_step_50729.pth"

gallery_parts = os.listdir(gallery_splits_dir)
query_parts = os.listdir(query_splits_dir)

for gallery_element in gallery_parts:
    for query_element in query_parts:
        # Sonuç dosya adı
        result_file = f"result_{gallery_element}_{query_element}.txt"
        cmd = (
            f"export PYTHONWARNINGS='ignore' ; "
            f"export PL_TQDM_REFRESH_RATE=0 ; "  
            f"tao model re_identification evaluate "
            f"-e {yaml_dir} "
            f"evaluate.checkpoint={model_dir} "
            f"evaluate.query_dataset={os.path.join(query_splits_dir, query_element)} "
            f"evaluate.test_dataset={os.path.join(gallery_splits_dir, gallery_element)} "
            f"re_ranking.re_ranking=True > {result_file} 2>&1"
        )
        print(f"Running: {cmd}")
        get_ipython().system(cmd)
```
